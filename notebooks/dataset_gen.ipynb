```python
import pandas as pd

data = []

age = 25
principal = 166000   
rate = 0.07         
years = 25

total_investment = 0
fd_value = 0

for year in range(1, years + 1):
    
   
    total_investment += principal
    
   
    fd_value = (fd_value + principal) * (1 + rate)
    
    interest_earned = fd_value - total_investment
    
    data.append([
        age,
        year,
        principal,
        round(fd_value, 2),
        round(interest_earned, 2),
        total_investment
    ])


df = pd.DataFrame(data, columns=[
    "Age",
    "Year",
    "Yearly_Investment",
    "FD_Value",
    "Interest_Earned",
    "Total_Investment"
])


df.to_csv(r"C:\Users\Suraj\Desktop\fd_dataset.csv", index=False)

print("FD dataset created successfully!")
```

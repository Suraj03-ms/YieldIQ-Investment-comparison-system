```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Suraj\Desktop\Gold_Yearly .csv")

df.columns = df.columns.str.strip().str.lower().str.replace('\n', ' ')

print("Columns:", df.columns)

df.rename(columns={
    'average closing price': 'price'
}, inplace=True)

df = df.sort_values('year').tail(25).reset_index(drop=True)

investment_per_year = 166000

data = []
total_investment = 0
total_grams = 0


for i in range(len(df)):
    year = i + 1
    price = df.loc[i, 'price']
    
    grams_bought = investment_per_year / price
    total_grams += grams_bought
    
    total_investment += investment_per_year
    
    current_value = total_grams * price
    
    data.append([
        25,
        year,
        investment_per_year,
        round(price, 2),
        round(total_grams, 2),
        round(current_value, 2),
        total_investment
    ])


gold_df = pd.DataFrame(data, columns=[
    "Age",
    "Year",
    "Investment",
    "Gold_Price",
    "Gold_Accumulated(g)",
    "Portfolio_Value",
    "Total_Investment"
])


gold_df.to_csv(r"C:\Users\Suraj\Desktop\gold_25years.csv", index=False)

print(" Gold dataset created successfully!")
```

```python
import pandas as pd
import numpy as np


df = pd.read_excel(r"C:\Users\Suraj\Desktop\mf_raw.xlsx")


df.columns = df.columns.str.strip().str.lower()


df['date'] = pd.to_datetime(df['date'], dayfirst=True)


df = df.sort_values('date')


df['nav'] = df['nav'].astype(float)


df['returns'] = df['nav'].pct_change()


avg_return = df['returns'].mean() * 252   

print("Estimated Annual Return:", avg_return)


years = 25
sip_yearly = 166000

total_investment = 0
portfolio_value = 0

data = []


for year in range(1, years + 1):
    
    total_investment += sip_yearly
    
    
    growth = np.random.normal(avg_return, 0.05)
    
    portfolio_value = (portfolio_value + sip_yearly) * (1 + growth)
    
    data.append([
        25,
        year,
        sip_yearly,
        total_investment,
        round(growth * 100, 2),
        round(portfolio_value, 2)
    ])


mf_df = pd.DataFrame(data, columns=[
    "Age",
    "Year",
    "Yearly_Investment",
    "Total_Investment",
    "Return_%",
    "Portfolio_Value"
])
mf_df.to_csv(r"C:\Users\Suraj\Desktop\mf_25years.csv", index=False)

print("✅ Mutual Fund 25-year dataset created successfully!")
```

```python
import pandas as pd

lic = pd.read_csv(r"C:\Users\Suraj\Desktop\lic_dataset.csv")
fd = pd.read_csv(r"C:\Users\Suraj\Desktop\fd_dataset.csv")
gold = pd.read_csv(r"C:\Users\Suraj\Desktop\gold_25years.csv")
mf = pd.read_csv(r"C:\Users\Suraj\Desktop\mf_25years.csv")
lic['Investment_Type'] = 'LIC'
fd['Investment_Type'] = 'FD'
gold['Investment_Type'] = 'Gold'
mf['Investment_Type'] = 'Mutual Fund'
if 'Final_Value' not in lic.columns:
    for col in lic.columns:
        if 'value' in col.lower():
            lic = lic.rename(columns={col: 'Final_Value'})

if 'Final_Value' not in fd.columns:
    fd = fd.rename(columns={'FD_Value': 'Final_Value'})
if 'Final_Value' not in gold.columns:
    for col in gold.columns:
        if 'value' in col.lower():
            gold = gold.rename(columns={col: 'Final_Value'})


if 'Final_Value' not in mf.columns:
    mf = mf.rename(columns={'Portfolio_Value': 'Final_Value'})


lic = lic[['Year', 'Total_Investment', 'Final_Value', 'Investment_Type']]
fd = fd[['Year', 'Total_Investment', 'Final_Value', 'Investment_Type']]
gold = gold[['Year', 'Total_Investment', 'Final_Value', 'Investment_Type']]
mf = mf[['Year', 'Total_Investment', 'Final_Value', 'Investment_Type']]


combined_df = pd.concat([lic, fd, gold, mf], ignore_index=True)


combined_df.to_csv(r"C:\Users\Suraj\Desktop\combined_dataset.csv", index=False)

print(" Combined dataset created successfully!")
print(combined_df.head())
```
